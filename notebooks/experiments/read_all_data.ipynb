```python
import numpy as np
import os

# Function to convert DataFrame row to JSON serializable format
def convert_row_to_serializable(row):
    row_dict = row.to_dict()
    for key, value in row_dict.items():
        if isinstance(value, (pd.Timestamp, pd.Timedelta)):
            row_dict[key] = str(value)
        elif isinstance(value, (np.bool_, np.integer, np.floating, np.ndarray)):
            row_dict[key] = value.item()
    return row_dict

```
